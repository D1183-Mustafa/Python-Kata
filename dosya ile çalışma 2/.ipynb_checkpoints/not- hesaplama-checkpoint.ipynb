```python
def not_hesaplama(n):
    
    m = n.split(",")
    isim = m[0]
    not1 = m[1]
    not2 = m[2]
    not3 = m[3]
    
    ortalama = int(not1) * 0.3 + int(not2) * 0.3 + int(not3) * 0.4
    
    if ortalama > 90:
        harf = "AA"
    elif 80 <= ortalama <= 90:
        harf = "BA"
    elif 70 <= ortalama < 80:
        harf = "BB"
    elif 60 <= ortalama < 70:
        harf = "CB"
    elif 50 <= ortalama < 60:
        harf = "CC"
    elif 40 <= ortalama < 50:
        harf = "CD"
    elif 30 <= ortalama < 40:
        harf = "DD"
    else:
        harf = "FF"
        
    return isim + "-->" + str(ortalama) + "-->" + harf + "\n"



with open("dosya.txt", "r", encoding= "utf-8") as dosya:
    notlar = []
    
    for i in dosya:
        notlar.append(not_hesaplama(i))
        
with open("notlar.txt", "w", encoding="utf-8") as dosya:
    dosya.writelines(notlar)
    
with open("notlar.txt", "r", encoding="utf-8") as dosya:
    print(dosya.read())
```
