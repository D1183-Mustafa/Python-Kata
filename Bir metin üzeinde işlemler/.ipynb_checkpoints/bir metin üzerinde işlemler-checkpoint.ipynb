```python
class Dosya():
    def __init__(self):
        
        with open("metin.txt","r",encoding="utf-8") as dosya:
            dosya_içeriği = dosya.read()
            
            kelimeler = dosya_içeriği.split()
            
            self.sade_kelimeler = []
            
            for i in kelimeler:
                i = i.strip("\n")
                i = i.strip(".")
                i = i.strip(" ")
                i = i.strip(",")
                self.sade_kelimeler.append(i)
                
    def tümkelimeler(self):
        kelimeler_kümesi = set()
        
        for i in self.sade_kelimeler:
            kelimeler_kümesi.add(i)
            
        print("Tüm kelimeler ....")
        
        for i in kelimeler_kümesi:
            print(i)
            print("*********************")
            
    def kelime_frekansı(self):
        kelime_sözlük = {}
        
        for i in self.sade_kelimeler:
            if i in kelime_sözlük:
                kelime_sözlük[i] += 1
            else:
                kelime_sözlük[i] = 1
                
                
        for i,j in kelime_sözlük.items():
            print("{} kelimesinden {} adet bulunmaktadır.".format(i,j))
            print("---------------------------")
            
        
```

```python
dosya = Dosya()
```

```python
dosya.kelime_frekansı()
```

```python
dosya.tümkelimeler()
```
